```python
# Project 1B: Library Book Management System

from tkinter import *
import sqlite3
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
```

```python
root= Tk()
root.geometry('1320x600')
root.title('A Library management Application!')
```

```python
def Database():
    global con, cur
    con = sqlite3.connect('library.db')
    cur = con.cursor()
    cur.execute('CREATE TABLE IF NOT EXISTS books(bookid INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, booktitle TEXT, author TEXT, year INTIGER, isbn TEXT)')

def GetValue(event):
    bookttl.delete(0, END)
    authr.delete(0, END)
    yr.delete(0, END)
    sbn.delete(0, END)
    
    row_id = tree.selection()[0]
    select = tree.set(row_id)
    bookttl.insert(0, select['Booktitle'])
    authr.insert(0, select['Author'])
    yr.insert(0, select['Year'])
    sbn.insert(0, select['ISBN'])

def add():
    if BOOKTITLE.get() == '' or AUTHOR.get() == '' or YEAR.get() == '' :
        txt_res.config(text = 'Please enter all the * feilds', fg='red')
    else:
        Database()
        cur.execute('INSERT INTO books(booktitle, author, year, isbn) VALUES (?,?,?,?)', (str(BOOKTITLE.get()), str(AUTHOR.get()), str(YEAR.get()), str(ISBN.get())))
        con.commit()
        BOOKTITLE.set('')
        AUTHOR.set('')
        YEAR.set('')
        ISBN.set('')
        cur.close()
        con.close()
        txt_res.config(text = 'Result inserted! ', fg= 'green')
        
def search():
    tree.delete(*tree.get_children())
    Database()
    cur.execute('SELECT * FROM books WHERE Booktitle = ?', (str(BOOKTITLE.get()),)) # Orders the data by booktitle in ascending order
    fetch1 = cur.fetchall()
    for data in fetch1:
        tree.insert('', 'end', values = (data[1],data[2],data[3],data[4]))
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    cur.close()
    txt_res.config(text='successfully fetched data! ', fg='blue')
    
def display():
    tree.delete(*tree.get_children())
    Database()
    cur.execute('SELECT * FROM books ORDER BY booktitle ASC') # Orders the data by booktitle in ascending order
    fetch = cur.fetchall()
    for data in fetch:
        tree.insert('', 'end', values = (data[1],data[2],data[3],data[4]))
    cur.close()
    txt_res.config(text='successfully fetched data! ', fg='blue')
    
def delete():
    
    Database() 
    cur.execute('DELETE FROM books WHERE Booktitle = ?', (str(BOOKTITLE.get()),))
    con.commit()
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    cur.close()
    txt_res.config(text='Issued the book! ', fg='blue')
    
def issue():
    Database() 
    cur.execute('DELETE FROM books WHERE Booktitle = ?', (str(BOOKTITLE.get()),))
    con.commit()
    BOOKTITLE.set('')
    AUTHOR.set('')
    YEAR.set('')
    ISBN.set('')
    cur.close()
    txt_res.config(text='successfully deleted data! ', fg='blue')
                
def Exit():
    result = tkMessageBox.askquestion('Do you want to exit? (y/n)', icon='warning')
    if result == 'yes':
        root.destroy()
        exit()
        
```

```python
#Variable declaration

BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR = IntVar()
ISBN = StringVar()
```

```python
# FRAMING

Top = Frame(root, width=400, height=50, bd=8, relief='flat')
Top.pack(side=TOP)

Left = Frame(root, width=400, height=500, bd=8, relief='flat')
Left.pack(side=LEFT)

Right = Frame(root, width=400, height=500, bd=8, relief='raise')
Right.pack(side=RIGHT)

Forms = Frame(Left, width=400, height=450, bd=8, relief='raise')
Forms.pack(side=TOP)

Buttons = Frame(Left, width=400, height=100, bd=8, relief='raise')
Buttons.pack(side=BOTTOM)
```

```python
# Labels

title = Label(Top, width=600, font=('arial', 24), text='.....Library Management Application.....')
title.pack()

booktitle = Label(Forms, text='Booktitle*: ', font=('arial', 16), bd=15)
booktitle.grid(row=0, stick='e')

author = Label(Forms, text='Author*: ', font=('arial', 16), bd=15)
author.grid(row=1, stick='e')

year = Label(Forms, text='Year*: ', font=('arial', 16), bd=15)
year.grid(row=2, stick='e')

isbn = Label(Forms, text='ISBN: ', font=('arial', 16), bd=15)
isbn.grid(row=3, stick='e')
```

```python
# Entry widgets

bookttl = Entry(Forms, textvariable=BOOKTITLE, width=30)
bookttl.grid(row=0, column=1)

authr = Entry(Forms, textvariable=AUTHOR, width=30)
authr.grid(row=1, column=1)

yr = Entry(Forms, textvariable=YEAR, width=30)
yr.grid(row=2, column=1)

sbn = Entry(Forms, textvariable=ISBN, width=30)
sbn.grid(row=3, column=1)
```

```python
# Buttons widget

btndisplay = Button(Buttons, width=10, text='Display', command=display)
btndisplay.pack(side=LEFT)

btnsearch = Button(Buttons, width=10, text='Search', command=search)
btnsearch.pack(side=LEFT)

btnadd = Button(Buttons, width=10, text='Add', command=add)
btnadd.pack(side=LEFT)

btnissue = Button(Buttons, width=10, text='Issue', command=issue)
btnissue.pack(side=LEFT)

btndelete = Button(Buttons, width=10, text='Delete', command=delete)
btndelete.pack(side=LEFT)

btnexit = Button(Buttons, width=10, text='Exit', command=Exit)
btnexit.pack(side=LEFT)

txt_res = Label(Buttons)
txt_res.pack()

# List widget

scrollbary = Scrollbar(Right, orient=VERTICAL)
scrollbarx = Scrollbar(Right, orient=HORIZONTAL)

tree = ttk.Treeview(Right, columns=('Booktitle', 'Author', 'Year', 'ISBN'))

scrollbary.config(command=tree.yview)
scrollbary.pack(side=RIGHT, fill=Y)

scrollbarx.config(command=tree.xview)
scrollbarx.pack(side=BOTTOM, fill=X)

tree.heading('Booktitle', text='Booktitle', anchor=W )
tree.heading('Author', text='Author', anchor=W )
tree.heading('Year', text='Year', anchor=W )
tree.heading('ISBN', text='ISBN', anchor=W )

tree.column('#0', stretch=NO, minwidth=0, width=0)
tree.column('#1', stretch=NO, minwidth=0, width=180)
tree.column('#2', stretch=NO, minwidth=0, width=120)
tree.column('#3', stretch=NO, minwidth=0, width=80)
tree.column('#4', stretch=NO, minwidth=0, width=100)
tree.pack()

# INTIALIZATION


tree.bind('<Double-Button-1>', GetValue)

if __name__ == '__main__':
    root.mainloop()
```
