```python
import pandas as pd
import numpy as np
```

```python
cust_det = pd.read_csv('.\pandas\customer_details.csv')

cust_det
```

```python
pol_det = pd.read_csv('.\pandas\customer_policy_details.csv')

pol_det
```

```python
cust_det.columns = ['customer_id', 'Gender', 'age', 'driving licence present', 'region code',
                    'previously insured', 'vehicle age', 'vehicle damage']
cust_det
```

```python
pol_det.columns = ['customer_id', 'annual premium(in Rs.)', 'sales channel code', 'vintage', 'response']

pol_det
```

```python
# count of null values

print(cust_det.isnull().sum(),'\n\n\n\n', pol_det.isnull().sum())
```

```python
# dropping null values for customer_id

cd1 = cust_det.dropna(subset = ['customer_id'])
cpd1 = pol_det.dropna(subset = ['customer_id'])

print(cd1.isna().sum(), '\n\n\n\n', cpd1.isna().sum())
```

```python
cd1[['age', 'region code']]=cd1[['age', 'region code']].fillna(cd1[['age', 'region code']].mean())
```

```python
cpd1['annual premium(in Rs.)']=cpd1['annual premium(in Rs.)'].fillna(cpd1['annual premium(in Rs.)'].mean())
cpd1['sales channel code']=cpd1['sales channel code'].fillna(cpd1['sales channel code'].mean())
cpd1['response']=cpd1['response'].fillna(cpd1['response'].mode()[0])
cpd1['vintage']=cpd1.vintage.fillna(cpd1['vintage'].mean())
```

```python
#replacing null values for categorical values by mode

cd1[['Gender','driving licence present', 'previously insured', 'vehicle age', 'vehicle damage']] = cd1[['Gender','driving licence present', 'previously insured', 'vehicle age', 'vehicle damage']].fillna(cd1[['Gender', 'driving licence present', 'previously insured', 'vehicle age', 'vehicle damage']].mode().iloc[0])
```

```python
# Replacing outlier values with mean values

cpd1.loc[(cpd1['annual premium(in Rs.)'] < 1952.375) , 'annual premium(in Rs.)' ] = cpd1['annual premium(in Rs.)'].mean()
cpd1.loc[(cpd1['annual premium(in Rs.)'] > 61855.375) , 'annual premium(in Rs.)' ] = cpd1['annual premium(in Rs.)'].mean()
```

```python
#Removing white spaces

cd1['Gender'] = cd1['Gender'].str.strip()
cd1['vehicle age'] = cd1['vehicle age'].str.strip()
cd1['vehicle damage'] = cd1['vehicle damage'].str.strip()

```

```python
#Case correction

cd1['Gender'] = cd1['Gender'].str.lower()
cd1['vehicle age'] = cd1['vehicle age'].str.lower()
cd1['vehicle damage'] = cd1['vehicle damage'].str.lower()
```

```python
# Dummies for nominal data

cd_dummie = cd1[['Gender','driving licence present', 'previously insured', 'vehicle age', 'vehicle damage']]
cpd_dummie = cpd1['response']
```

```python
# Dropping duplicates

cd1 = cd1.drop_duplicates()
```

```python
cpd1 = cpd1.drop_duplicates()
```

```python
master_table = pd.merge(cd1,cpd1, on='customer_id')
master_table
```

```python
# Average annual premium by gender

master_table.groupby(['Gender']).mean()['annual premium(in Rs.)']
```

```python
# Average annual premium by age

master_table.groupby(['age']).mean()['annual premium(in Rs.)']
```

```python
#checking if data is balanced between genders

male, female=0,0

for i in master_table['Gender']:
    
    if i == 'male':
        male += 1
    elif i == 'female':
        female += 1
        
print(male, female)
    
# Data is not balanced between genders
```

```python
# Vehicle age wise average annual premium

master_table.groupby(['vehicle age']).mean()['annual premium(in Rs.)']
```
